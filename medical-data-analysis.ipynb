```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv("medical_data.csv")

# Display first few rows of the dataset
print("Dataset Preview:")
print(data.head())

# Calculate average vitals
average_vitals = data.mean(numeric_only=True)
print("\nAverage Vitals:")
print(average_vitals)

# Statistical summary of the data
print("\nStatistical Summary:")
print(data.describe())

# Calculate abnormal readings
# Thresholds for high and low blood pressure
bp_high_threshold = 140
bp_low_threshold = 90

# Calculation of abnormal readings for Blood Pressure
abnormal_data_bp = data[
    (data["Blood Pressure"] > bp_high_threshold) |
    (data["Blood Pressure"] < bp_low_threshold)
]

# Print abnormal readings
print("Abnormal Readings (Blood Pressure):")
print(abnormal_data_bp)

# Visualizing the distribution of blood pressure values 
# With a Histogram
plt.figure(figsize=(8,5))
plt.hist(data["Blood Pressure"], bins=10, color='blue', alpha=0.7)
plt.xlabel("Blood Pressure (mmHg)")
plt.ylabel("Frequency")
plt.title("Distribution of Blood Pressure Readings")
plt.grid(True)
plt.show()

# Visualizing blood pressure trends over time
# With a line plot
data["Date"] = pd.to_datetime(data["Date"])  # Convert to datetime
data = data.sort_values("Date")  # Sort by date

plt.figure(figsize=(10,5))
plt.plot(data["Date"], data["Blood Pressure"], marker='o', linestyle='-', color='red', label="Blood Pressure")
plt.xlabel("Date")
plt.ylabel("Blood Pressure (mmHg)")
plt.title("Blood Pressure Trends Over Time")
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()

# Scatter Plot for Blood Pressure vs. Heart Rate
plt.figure(figsize=(8,5))
plt.scatter(data["Blood Pressure"], data["Heart Rate"], color='green', alpha=0.5)
plt.xlabel("Blood Pressure (mmHg)")
plt.ylabel("Heart Rate (bpm)")
plt.title("Blood Pressure vs. Heart Rate")
plt.grid(True)
plt.show()

# Box Plot to Identify Outliers
plt.figure(figsize=(8,5))
plt.boxplot(data["Blood Pressure"], vert=False, patch_artist=True)
plt.xlabel("Blood Pressure (mmHg)")
plt.title("Blood Pressure Outlier Detection")
plt.show()
```
